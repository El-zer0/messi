```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['Exam_Score'],axis=1)
y=data['Exam_Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaller=MinMaxScaler()
ScallerCols=['Study Time','Absences']

x_train[ScallerCols]=scaller.fit_transform(x_train[ScallerCols])
x_test[ScallerCols]=scaller.transform(x_test[ScallerCols])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE ,r2_score as R2 , mean_absolute_error as MAE
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
y_pred_LR=LR.predict(x_test)
print(MSE(y_test,y_pred_LR))
print(R2(y_test,y_pred_LR))
print(MAE(y_test,y_pred_LR))
jb.dump(LR,'LR.pkl')
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
y_pred_RFR=RFR.predict(x_test)
print(MSE(y_test,y_pred_RFR))
print(R2(y_test,y_pred_RFR))
print(MAE(y_test,y_pred_RFR))
jb.dump(RFR,'RFR.pkl')
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
y_pred_DTR=DTR.predict(x_test)
print(MSE(y_test,y_pred_DTR))
print(R2(y_test,y_pred_DTR))
print(MAE(y_test,y_pred_DTR))
jb.dump(DTR,'DTR.pkl')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=y_test,y=y_pred_RFR,color='Red')
sns.scatterplot(x=y_test,y=y_pred_LR,color='Blue')
sns.scatterplot(x=y_test,y=y_pred_DTR,color='Green')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
Grid_Pram={'max_depth': [3, 5, 10, None], 'max_leaf_nodes': [2,5,7]}
Grid_Search=GridSearchCV(DecisionTreeRegressor(),Grid_Pram,cv=5)
Grid_Search.fit(x_train,y_train)
print('Grid Scoure: ',Grid_Search.best_params_)
```

```python
y_prediction_Grid=Grid_Search.best_estimator_.predict(x_test)
print(MSE(y_test,y_prediction_Grid))
print(R2(y_test,y_prediction_Grid))
print(MAE(y_test,y_prediction_Grid))
```

```python
RSC_Pram={'n_estimators': [5,100,200],'max_depth': [3, 5, 10]}
RSC_Model=RandomizedSearchCV(RandomForestRegressor(),RSC_Pram,scoring='neg_mean_squared_error',random_state=42)
RSC_Model.fit(x_train,y_train)
print('RSC Scoure: ',RSC_Model.best_params_)
```

```python
y_prediction_RSC=RSC_Model.best_estimator_.predict(x_test)
print(MSE(y_test,y_prediction_RSC))
print(R2(y_test,y_prediction_RSC))
print(MAE(y_test,y_prediction_RSC))
```
